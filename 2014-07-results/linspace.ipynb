# Linspace

```python
import benchmark_analysis_utils as bau
df = bau.load_results_file("results_1405803619.csv").sort()
```

## Small

```python
df.loc[('small', 'esb', 'linspace')].sort('compress')['compress'].plot(kind='barh', figsize=(15,40))
```

* Differences not as pronounced as with the arange
* bloscpack with some codecs ahead of the pack
* NPZ and ZFile are slow

```python
df.loc[('small', 'esb', 'linspace')].sort('decompress')['decompress'].plot(kind='barh', figsize=(15,40))
```

* Bloscpack with various codecs slightly better than npy
* NPZ and ZFile are slow again

```python
df.loc[('small', 'esb', 'linspace')].sort('dc_no_cache')['dc_no_cache'].plot(kind='barh', figsize=(15,40))
```

* On a cold cache, ALL the bloscpacks are faster than NPY

```python
df.loc[('small', 'esb', 'linspace')].sort('ratio')['ratio'].plot(kind='barh', figsize=(15,40))
```

* Just like arange bloscpack with zlib and level 9 is great, only 60x but still

## Large

```python
df.loc[('large', 'esb', 'linspace')].sort('compress')['compress'].plot(kind='barh', figsize=(15,40))
```

* Again ZFile and NPZ are so far behind, we need to reduce the graph

```python
codecs = set(df.index.levels[-2]).difference(set(('zfile', 'npz')))
df.loc[('large', 'esb', 'linspace', tuple(codecs))].sort('compress')['compress'].plot(kind='barh', figsize=(15,40))
```

* At this size, any compression is better than plain
* Bloscpack with lz4 and blosclz at level 7 and 9 around 10x faster

```python
df.loc[('large', 'esb', 'linspace')].sort('decompress')['decompress'].plot(kind='barh', figsize=(15,40))
```

* Some of the bloscpack codecs are faster than NPY

```python
df.loc[('large', 'esb', 'linspace')].sort('dc_no_cache')['dc_no_cache'].plot(kind='barh', figsize=(15,40))
```

* Mostly all faster than plain, with the usual exceptions
* Interestingly bloscpack with lz4hc and level 3 wins here

```python
df.loc[('large', 'esb', 'linspace')].sort('ratio')['ratio'].plot(kind='barh', figsize=(15,40))
```

* Again the bloscpack with zlib has an amazing ratio
