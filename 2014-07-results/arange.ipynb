# Arange

```python
import benchmark_analysis_utils as bau
df = bau.load_results_file("results_1405803619.csv").sort()
```

## Small

```python
df.loc[('small', 'esb', 'arange')].sort('compress')['compress'].plot(kind='barh', figsize=(15,40))
```

We can see:
    
* Zfile and NPZ are much slower
* Bloscpack with some codecs is fastest
* Many faster than NPY, not by much, but faster

```python
df.loc[('small', 'esb', 'arange')].sort('decompress')['decompress'].plot(kind='barh', figsize=(15,40))
```

* Differences are not pronounced
* Codecs are close together
* Some are actully faster than NPY, even with hot FS cache

```python
df.loc[('small', 'esb', 'arange')].sort('dc_no_cache')['dc_no_cache'].plot(kind='barh', figsize=(15,40))
```

* tables with zlib is loosing
* also tables is completely at the top, perhaps initialization overhead
* bloscpack_lz4, 5 is almost twice as fast as NPY

```python
df.loc[('small', 'esb', 'arange')].sort('ratio')['ratio'].plot(kind='barh', figsize=(15,40))
```

* NPY and tables have a ratio of 1
* zfile and npz have pretty bad ratios compared to the rest, presumably due to missing shuffle
* snappy doesn't have a competetive ratio
* bloscpack_zlib gives a compression ratio of almost 300

## Large

```python
df.loc[('large', 'esb', 'arange')].sort('compress')['compress'].plot(kind='barh', figsize=(15,40))
```

* zfile just sucks on this dataset
* Many codecs beat NPY with this size

Let's take the costly ones out of the equation

```python
codecs = set(df.index.levels[-2]).difference(set(('zfile', 'npz')))
df.loc[('large', 'esb', 'arange', tuple(codecs))].sort('compress')['compress'].plot(kind='barh', figsize=(15,40))
```

* Okay great, we can see how certain combinations/codecs are almost 10x faster
* in fact, all of the remaining ones are faster than the uncompressed formats

```python
df.loc[('large', 'esb', 'arange')].sort('decompress')['decompress'].plot(kind='barh', figsize=(15,40))
```

* No suprises, as with the small dataset
* Remember however that this is with a hot FS cache

```python
df.loc[('large', 'esb', 'arange')].sort('dc_no_cache')['dc_no_cache'].plot(kind='barh', figsize=(15,40))
```

* At this size, any compression will beneficial over plain formats

```python
df.loc[('large', 'esb', 'arange')].sort('ratio')['ratio'].plot(kind='barh', figsize=(15,40))
```

* bloscpack_zlib and tables_zlib continue to win this time even by > 350x
